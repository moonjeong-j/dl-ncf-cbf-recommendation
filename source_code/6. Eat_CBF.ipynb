```python
result_df =pd.DataFrame()

for user_id in tqdm(test_userids):
  user_record_df = train_df.loc[train_df['u_id']==int(user_id),:] # (n,3) 

  user_sim_df= cs_p_df.loc[user_record_df['p_id']] #(n,1730). #1730개......
  user_rating_df = user_record_df[['rating']] #(n,1)
  sim_sum = np.sum(user_sim_df.T.to_numpy(),-1)


  prediction =np.matmul(user_sim_df.T.to_numpy() , user_rating_df.to_numpy()).flatten() /(sim_sum+1)
  
  prediction_df = pd.DataFrame(prediction , index = cs_p_df.index).reset_index()
  prediction_df.columns=['p_id', 'pred_rating']
  prediction_df = prediction_df[['p_id', 'pred_rating']][prediction_df['p_id'].isin(test_df[test_df['u_id'] == user_id]['p_id'].values)]

  temp_df = prediction_df.merge(test_df[test_df['u_id'] == user_id], on='p_id')
  result_df = pd.concat([result_df, temp_df], axis=0)
```
